```python
# Display the count of missing (null) values for each column in the 2022 dataset
display(hdi_2022_df.isnull().sum())

# Print the total number of missing values in the 2022 dataset
print("Total Count:")
print(hdi_2022_df.isnull().sum().sum())
```

```python
# Select columns with categorical (object) data types from the 2022 dataset
categorical_columns = hdi_2022_df.select_dtypes(include=['object']).columns

# Select columns with numerical data types from the 2022 dataset
numerical_columns = hdi_2022_df.select_dtypes(exclude=['object']).columns
```

Replacing the special characters with Nan for consistent calcuation.

```python
# Replace common placeholders for missing values (such as "–", "-", "NA", "n/a") with NaN in the 2022 dataset
hdi_2022_df.replace(["–", "-", "NA", "n/a"], np.nan, inplace=True)
```

Appropriate respective conversion to numeric type and replacement with Nan if missing or malformed data.

```python
# Convert all columns with numerical data to numeric type, coercing errors to NaN in the 2022 dataset
for col in numerical_columns:
    hdi_2022_df[col] = pd.to_numeric(hdi_2022_df[col], errors='coerce')
```

```python
# Remove leading and trailing whitespaces from the 'country' column in the 2022 dataset
hdi_2022_df['country'] = hdi_2022_df['country'].str.strip()
```

```python
# Remove duplicate rows from the 2022 dataset
hdi_2022_df.drop_duplicates(inplace=True)
```

```python
# Drop rows where the 'hdi' column has missing (NaN) values in the 2022 dataset
hdi_2022_df = hdi_2022_df.dropna(subset=['hdi'])
```

Imputing with mean if missing values exceed 5% in overall else dropping the rows.

```python
# Handle missing values in numerical columns (except 'hdi') in the 2022 dataset
for col in numerical_columns:
    if col == 'hdi':  # Skip the 'hdi' column
        continue
    # Calculate the percentage of missing values for the column
    missing_pct = hdi_2022_df[col].isnull().sum() / len(hdi_2022_df) * 100
    # If missing values are more than 5%, fill them with the column's mean
    if missing_pct > 5:
        hdi_2022_df[col] = hdi_2022_df[col].fillna(hdi_2022_df[col].mean())
    # If missing values are less than or equal to 5%, drop rows with missing values
    else:
        hdi_2022_df = hdi_2022_df.dropna(subset=[col])
```

```python
# Display the first few rows of the cleaned 2022 dataset
print("Cleaned Data:")
hdi_2022_df.head()
```

6. Adding HDI Category Column:

```python
# Function to categorize HDI based on specified thresholds
def hdi_categories(hdi):
    # If HDI is less than 0.550, categorize as 'Low'
    if hdi < 0.550:
        return 'Low'
    # If HDI is between 0.550 and 0.699 (inclusive), categorize as 'Medium'
    elif 0.550 <= hdi <= 0.699:
        return 'Medium'
    # If HDI is between 0.700 and 0.799 (inclusive), categorize as 'High'
    elif 0.700 <= hdi <= 0.799:
        return 'High'
    # If HDI is above 0.799, categorize as 'Very High'
    else:
        return 'Very High'
```

```python
# Apply the 'hdi_categories' function to the 'hdi' column to categorize each country's HDI
hdi_2022_df['HDI_Category'] = hdi_2022_df['hdi'].apply(hdi_categories)

# Print the count of countries in each HDI category
print(hdi_2022_df['HDI_Category'].value_counts())

print()  # Add a blank line for separation

# Print the first 10 rows with the country, HDI, and its corresponding HDI category
print("HDI with its HDI Category Comparision\n")
hdi_2022_df[['country', 'hdi', 'HDI_Category']].head(10)
```

```python
# Filter the dataset for the year 2022 and create a new DataFrame
hdi_south_asia_2022 = hdi_south_asia_df[hdi_south_asia_df['year'] == 2022].copy()
```

```python
# Replace common placeholders for missing values (such as "–", "-", "NA", "n/a") with NaN
hdi_south_asia_2022.replace(["–", "-", "NA", "n/a"], np.nan, inplace=True)

# Define the columns that are numeric and need conversion
numeric_cols = ['hdi', 'life_expectancy', 'gross_inc_percap']

# Convert the numeric columns to numeric type, coercing errors to NaN
for col in numeric_cols:
    hdi_south_asia_2022[col] = pd.to_numeric(hdi_south_asia_2022[col], errors='coerce')

# Remove duplicate rows from the dataset
hdi_south_asia_2022.drop_duplicates(inplace=True)

# Handle missing values in the numeric columns
for col in numeric_cols:
    # Calculate the percentage of missing values for the current column
    missing_pct = hdi_south_asia_2022[col].isnull().sum() / len(hdi_south_asia_2022) * 100

    # If more than 5% of the values are missing, fill with the column's mean
    if missing_pct > 5:
        hdi_south_asia_2022[col] = hdi_south_asia_2022[col].fillna(hdi_south_asia_2022[col].mean())
    # If less than or equal to 5% of the values are missing, drop rows with missing values in this column
    else:
        hdi_south_asia_2022 = hdi_south_asia_2022.dropna(subset=[col])
```

**3. Outlier Detection:**

```python
# Calculate the first and third quartiles (Q1 and Q3) for HDI
Q1_hdi = hdi_south_asia_2022['hdi'].quantile(0.25)
Q3_hdi = hdi_south_asia_2022['hdi'].quantile(0.75)

# Calculate the Interquartile Range (IQR) for HDI
IQR_hdi = Q3_hdi - Q1_hdi

# Determine the lower and upper bounds for HDI outliers
lower_hdi = Q1_hdi - 1.5 * IQR_hdi
upper_hdi = Q3_hdi + 1.5 * IQR_hdi

# Calculate the first and third quartiles (Q1 and Q3) for GNI per Capita
Q1_gni = hdi_south_asia_2022['gross_inc_percap'].quantile(0.25)
Q3_gni = hdi_south_asia_2022['gross_inc_percap'].quantile(0.75)

# Calculate the Interquartile Range (IQR) for GNI per Capita
IQR_gni = Q3_gni - Q1_gni

# Determine the lower and upper bounds for GNI outliers
lower_gni = Q1_gni - 1.5 * IQR_gni
upper_gni = Q3_gni + 1.5 * IQR_gni
```

```python
# Identify outliers based on the HDI and GNI per Capita using the previously calculated bounds
hdi_south_asia_2022['Outlier'] = (
    (hdi_south_asia_2022['hdi'] < lower_hdi) | (hdi_south_asia_2022['hdi'] > upper_hdi) |
    (hdi_south_asia_2022['gross_inc_percap'] < lower_gni) | (hdi_south_asia_2022['gross_inc_percap'] > upper_gni)
)

# Print the countries with their HDI, GNI per Capita, and whether they are outliers
print("Outlier Detection (HDI & GNI per Capita) for South Asia, 2022:\n")
print(hdi_south_asia_2022[['country','hdi','gross_inc_percap','Outlier']])
```

```python
# Create a scatter plot to visualize the relationship between GNI per Capita and HDI, with outliers highlighted
plt.figure(figsize=(8,6))  # Set the figure size

# Create a scatter plot, coloring points by the 'Outlier' status (red for outliers, blue for non-outliers)
sns.scatterplot(data=hdi_south_asia_2022, x='gross_inc_percap', y='hdi', hue='Outlier', palette={False: 'blue', True: 'red'})

# Add a title and labels for the axes
plt.title("HDI vs GNI per Capita (South Asia, 2022)")
plt.xlabel("GNI per Capita")
plt.ylabel("Human Development Index (HDI)")

# Add a legend to distinguish outliers from non-outliers
plt.legend(title="Outlier")
plt.savefig('GNI_per_capita_outlier_detection.png', dpi=300)
# Display the plot
plt.show()
```

Using the 1.5*IQR rule, no countries in 2022 are classified as outliers in either HDI or GNI per capita because it lies between the lower and upper bound for both.

**7. Outlier Detection:**

```python
# Function to detect outliers based on the Interquartile Range (IQR)
def detect_outliers(df, col):
    # Calculate the first (Q1) and third (Q3) quartiles of the specified column
    Q1 = df[col].quantile(0.25)  # 25th percentile (Q1)
    Q3 = df[col].quantile(0.75)  # 75th percentile (Q3)
    
    # Calculate the Interquartile Range (IQR)
    IQR = Q3 - Q1  # IQR = Q3 - Q1
    
    # Define the lower and upper bounds for outlier detection
    lower = Q1 - 1.5 * IQR  # Any value below this is considered an outlier
    upper = Q3 + 1.5 * IQR  # Any value above this is considered an outlier
    
    # Identify the outliers by filtering rows where the value is either below the lower bound or above the upper bound
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    
    # Return the rows that are outliers
    return outliers
```

```python
# South Asia: Detect and display outliers for HDI and GNI per Capita

# Detect outliers in HDI column for South Asia
outlier_hdi_south = detect_outliers(hdi_south_asia_2020_2022, 'hdi')  # Using the detect_outliers function for HDI
# Detect outliers in GNI per Capita column for South Asia
outlier_gni_south = detect_outliers(hdi_south_asia_2020_2022, 'gross_inc_percap')  # Using the detect_outliers function for GNI per Capita

# Print the detected outliers for HDI
print("South Asia HDI Outliers:")
print(outlier_hdi_south[['country','hdi']])  # Display country and HDI for outliers

# Print the detected outliers for GNI per Capita
print("\nSouth Asia GNI per Capita Outliers:")
print(outlier_gni_south[['country','gross_inc_percap']])  # Display country and GNI for outliers
```

```python
# South Asia: Scatter plot showing HDI vs GNI per Capita with outliers highlighted

plt.figure(figsize=(8,6))  # Set the figure size

# Plot normal countries (non-outliers) with blue dots
plt.scatter(hdi_south_asia_2020_2022['gross_inc_percap'], hdi_south_asia_2020_2022['hdi'], color='blue', label='Normal Countries')

# Plot HDI outliers with red dots
plt.scatter(outlier_hdi_south['gross_inc_percap'], outlier_hdi_south['hdi'], color='red', label='HDI Outliers')

# Plot GNI per Capita outliers with orange dots
plt.scatter(outlier_gni_south['gross_inc_percap'], outlier_gni_south['hdi'], color='orange', label='GNI Outliers')

# Labels and title
plt.xlabel("GNI per Capita")  # X-axis label
plt.ylabel("HDI")  # Y-axis label
plt.title("South Asia: HDI vs GNI per Capita")  # Title of the plot

# Display legend to differentiate normal countries and outliers
plt.legend()
plt.savefig('HDI_vs_GNI_per_capita_outlier_SouthAsia.png', dpi=300)
# Display the plot
plt.show()
```
